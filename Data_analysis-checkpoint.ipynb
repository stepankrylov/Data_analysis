```python
import pandas as pd
import os
```

```python
count_top3 = ['1880']
list_data_frame = []
for item in count_top3:
    names = pd.read_csv(f'names\yob{item}.txt', names=['Name', 'Gender', 'Count'])
    names['Year'] = str(item)
    list_data_frame.append(names)
df_res_1 = pd.concat(list_data_frame)
df_res_1
```

```python
pd.set_option('display.max_rows', 10)
```

```python
df_res_1.groupby(['Name', 'Gender']).sum().sort_values('Count', ascending=False)[:3]
```
